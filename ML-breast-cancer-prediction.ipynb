# MACHINE LEARNING MODEL ANALYSIS

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
df.head()
```

```python
df['diagnosis'].unique()    #so now we see that instead of M and B ,we see 1 and 0.
```

```python
X = df.drop('diagnosis', axis=1)   #Apart from "diagnosis" column all are independent variables,so take the rest as "X".
X.head()                           #We see here "diagnosis" column is dropped
```

```python
y = df['diagnosis']       #This is our dependent variable ,so we store it in "Y".
y.head()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)  #Here 0.3 means to give 30% space to "test" part, and 70% to "train" part
```

```python
print(df.shape)
print(X_train.shape)          #upper left
print(X_test.shape)          #lower left
print(y_train.shape)          #upper right
print(y_test.shape)          #lower right
```

X_train.head(1)

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()                               #Need to create a object of "StandardScaler" function/library
X_train = ss.fit_transform(X_train)         #Since this is train data therefore we need to fit and transform/scale
X_test = ss.transform(X_test)             #Whereas here we don't need to fit, we just need to transform/scale
```

```python
X_train
```

# Machine Learning Models

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()                            #creating a "LogisticRegression()" object.
lr.fit(X_train, y_train)                           #Here we are fitting the 'X' and 'Y' training data into ML model
```

```python
y_pred = lr.predict(X_test)             #Here we are predicting the result for test input 'X' using Logistic Regression. 
```

```python
y_pred
```

```python
np.array(y_test)                  #just converted the "y_test" column into a array using "numpy" library function
```

```python
#To check accuracy between "y_pred" and "y_test" we use "accuracy_score" function of "sklearn.metrics" library

from sklearn.metrics import accuracy_score
lr_acc = accuracy_score(y_test, y_pred)
print(lr_acc)
```

```python
# For comparing the results of 4 algorithms we could list them in a "dataframe" one below the other.
# So we would save it in a variable - 'results'
# Dataframes are present in "panda" library

results = pd.DataFrame()
results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

## Decision Tree Classifier (more powerful than logistic regression)

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()                   #creating a "DecisionTreeClassifier() " object.
dtc.fit(X_train, y_train)                       #Here we are fitting the 'X' and 'Y' training data into ML model
```

```python
y_pred = dtc.predict(X_test)             #Here we are predicting the result for test input 'X' using Decision Tree Classifier. 
y_pred
```

```python
#To check accuracy between "y_pred" and "y_test" we use "accuracy_score" function of "sklearn.metrics" library

from sklearn.metrics import accuracy_score
dtc_acc = accuracy_score(y_test, y_pred)
print(dtc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision tree Classifier Method'], 'Accuracy':[dtc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
rfc_acc = accuracy_score(y_test, y_pred)
print(rfc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Random Forest Classifier Method'], 'Accuracy':[rfc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

## Support Vector Classifier 

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
svc_acc = accuracy_score(y_test, y_pred)
print(svc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Support Vector Classifier Method'], 'Accuracy':[svc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```
